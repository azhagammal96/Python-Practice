```python
#PICK WORDS
import numpy as np
import random

words = open("sowpods.txt", "r")
wordlist = [line.split() for line in words.readlines()]
choice = random.choice(wordlist)



len_str = len(choice[0])
lines = []

for i in range(len_str):
    lines.append("_")


def game():
    global counter
    counter = 0
    while  True:
        for i in range(100):
            ipt = input("Enter a letter: \n")

            if ipt in choice[0]:
                indexes = [x for x, v in enumerate(choice[0]) if v == ipt]
                for i in range(len(indexes)):
                    lines.pop(indexes[i])
                    lines.insert(indexes[i], ipt)
                    print(str(lines)+"\n")               
                            
            else:
                print("Incorrect guess, try again\n")
                counter += 1
            if "_" not in lines:
                return False

def main():
    print("Welcome to the hangman game. You have 100 turns to guess the word.Start guessing!\n")
    game()
    print("You win!\n")
    print("Number of incorrect guesses is: " +str(counter) + "\n")
    again = input("Would you like to try again?: \n")
    if again == "Y" or again == "yes" or again == "y":
        game()
    else:
        print("Have a good day\n")

main()
```
